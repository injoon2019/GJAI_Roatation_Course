### 데이터 시각화 테스트

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, ls='--') #line style '-', '--', '-.', ':'
plt.show()
```

```python
#Marker
plt.plot(x, y, ls=":", marker='o')  #marker o, v, ^, <, >, s, p, 1, 2, 3, 4, *, h, H, +, D, -
plt.show()
```

```python
#Line weight
plt.plot(x, y, lw = '5')
plt.show()
```

```python
#Line Color
plt.plot(x, y, color='r')
plt.show()
```

```python
plt.plot(x, y, marker='o', markerfacecolor='g', markersize=10, markeredgewidth=5) #mfc, ms, mew
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]

#선스타일 점선, 마커 점, 
plt.plot(x, y, ls="--", marker="o", markerfacecolor='g')
plt.show()
```

### Matplotlib 연습 01


```python
drive_path = r"c:/Apps/temp/"
data = pd.read_csv(drive_path+"Traffic_Accident_2017.csv", encoding="euc-kr")
data
```

```python
# 요일별 사고 건수
t = data["요일"].value_counts()
#value 순서대로 나온 것을 원하는 index 순서대로 정리하고 싶을 떄
t = t[['월', '화', '수', '목', '금', '토', '일']]
t
```

```python
x = range(7)
plt.plot(x, t)
plt.show()
```

```python
# x의 tick 바꾸기
# y의 limm 바꾸기 
x = range(7)
plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])
plt.xlabel("요일")
plt.ylabel("사고 건수")
plt.bar(x, t)
plt.ylim(500, 620)
plt.title("요일별 교통사고 사망 건수")
plt.show()
```

```python
## 차vs차 사고 다발지역 (발생지 시도 시각화) : 발생지시도, 사고유형_대분류
data.T

#data[["발생지시도", "사고유형_대분류"]]
#t = data['사고유형_대분류'][data['사고유형_대분류']=='차대차']
t = data[["발생지시도", "사고유형_대분류"]][data['사고유형_대분류']=='차대차']
print(t)

t1 = t['발생지시도'].value_counts()
y = t1[['서울', '대구', '부산', '광주', '경기', '제주']]
x = range(6)
plt.title('차 vs 차 사고 다발지역')
plt.xticks(x, ['서울', '대구', '부산', '광주', '경기', '제주'])
plt.bar(x, y)
plt.show()
```

### Pandas 시각화 도구

```python
import pandas as pd

df= pd.read_excel(drive_path+"남북한발전전력량.xlsx")
south = df.loc[0]
north = df.loc[5]

frames = [south, north]
result = pd.concat(frames, axis=1)
result = pd.DataFrame(result)
result = result.T
result
```

```python
df_ns= result.iloc[:, 2:]
df_ns.index=['한국', '북한']
df_ns=df_ns.T
plt.plot(df_ns)
plt.show()
```

```python
plt.plot(kind='bar')
plt.show()
```

```python
df_ns.plot(kind='hist')
```
