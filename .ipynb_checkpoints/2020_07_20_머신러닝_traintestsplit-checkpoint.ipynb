```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Iris daa loading
df = sns.load_dataset('iris')
df.shape
type(df)
df.head()
```

```python
#데이터 확인
X = df.iloc[:, :4]  #학습용 데이터 x 가져오기
y = df.iloc[:,-1]  #y 가져오기
```

```python
# 훈련 데이터와 테스트 데이터 준비
from sklearn.model_selection import train_test_split

#2차원 배열(행렬)의 수학 표기방식 대문자 X
#1치원 배열(벡터)의 수학 표기방식 소문자 Y
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0) #random_state = 0은 seed값을 주는 것이다

#75% : 25%
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#데이터 조사
#산점도 행렬: 3개 이상의 특성을 표현
#4개의 특성을 갖는 붓꽃
#대각원소 자리에 각 변수별 히스토그램

sns.pairplot(df, diag_kind="kde", hue="species", palette="bright")
plt.show()
```

공통

```python
#머신러닝 모델
#k-최근접 이웃 알고리즘
#훈련 데이터에서 새로운 데이터 포인트에 가장 가까운 'K개' 이웃을 찾는다

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 1)
model.fit(X_train, y_train)
```

```python
# 예측하기
X_new = np.array( [[5, 2.9, 1, 0.2]] )
X_new.shape

pre = model.predict(X_new)
print("예측:", pre)
```

```python
#모델 평가
#테스트 세트 이용
from sklearn import metrics

pre = model.predict(X_test)
ac_score = metrics.accuracy_score(y_test, pre)
print("정답률", ac_score)
```

### Regression

```python
#영화에 대한 3개의 특성
X_train = [
    [0.5, 0.2, 0.1],
    [0.9, 0.7, 0.3],
    [0.4, 0.5, 0.7]
]

# 영화에 대한 별점
y_train = [5.0, 6.8, 9.0]
```

```python
X_test = [
  [0.2, 0.1, 0.7],
  [0.4, 0.7, 0.6],
  [0.5, 0.8, 0.1]
]
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split


#이웃의 수를 3으로 하여 모델으 객체를 만듭니다
#단순한 평균을 적용('uniform')
reg = KNeighborsRegressor(n_neighbors=3, weights="distance") #가중 평균을 적용

#훈련 데이터와 타깃을 사용하여 모델을 학습시킵니다
reg.fit(X_train, y_train)

#테스트 세트에 대해 예측을 합니다
print("테스트 세트 예측:\n{}".format(reg.predict(X_test)))
type(X_train)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
import mglearn

X, y = mglearn.datasets.make_wave(n_samples = 40)

# wave 데이터 셋을 훈련 세트와 테스트 세트로 나눕니다
X_train, X_test, y_train, y_test =  train_test_split(X, y, random_state=0)

#이웃의 수를 3으로 하여 모델으 객체를 만듭니다
#단순한 평균을 적용('uniform')
reg = KNeighborsRegressor(n_neighbors=3)

#훈련 데이터와 타깃을 사용하여 모델을 학습시킵니다
reg.fit(X_train, y_train)

#테스트 세트에 대해 예측을 합니다
print("테스트 세트 예측:\n{}".format(reg.predict(X_test)))
type(X_train)
```
