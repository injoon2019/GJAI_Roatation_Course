```python
my_counts = df.region.value_counts()
my_counts.plot.bar(color="orange", alpha=0.3)
my_counts
```

```python
df2 = df[df.region=="east"]
df2.shape[0]
```

```python
df3 = df[df.region=="west"]
df3.shape[0]
```

```python
#위의 것을 다른 방식으로 하는 방법
df4 = pd.DataFrame( {"Counts": [df2.shape[0], df3.shape[0]]}, index=["East", "West"])
```

```python
df4.plot.bar(color="orange", alpha=0.5)
plt.show()
```

```python
df2.pm25.plot.hist(bins=30, color="blue", alpha=0.3, label="East")    #df2.loc[:, "pm25"]  이렇게해도된다
df3.pm25.plot.hist(bins=30, color="red", alpha=0.3, label="West")
plt.legend()
plt.show()
```

```python
df2.plot.scatter(x="latitude", y="pm25", color="blue", alpha=0.3, marker="o", s=20)
plt.show()

#동부 자료
```

```python
df3.plot.scatter(x="latitude", y="pm25", color="blue", alpha=0.3, marker="o", s=20)
plt.show()

#서부 자료
```

```python
#산점도는 bar 그래프와는 달리 overlay하려면 이 base를 써야한다. ax
base = df2.plot.scatter(x="latitude", y="pm25", color="blue", alpha=0.3, marker="o", s=20, label="East")
df3.plot.scatter(x="latitude", y="pm25", color="red", alpha=0.3, marker="o", s=20, ax=base, label="West")
plt.show()     
```

```python
#lat과 long 을 이용해서 지도
df.plot.scatter(x="longitude", y="latitude", color='green', alpha=0.5, marker="o", s=15)
plt.show()
```

```python
pd.plotting.scatter_matrix(df, color="green", alpha=0.3, marker=".")
plt.show()
```

```python
#특정 column을 제외하고 그리고 싶을 때 
pd.plotting.scatter_matrix(df.drop(columns="latitude"), color="green", alpha=0.3, marker=".")
plt.show()
```
