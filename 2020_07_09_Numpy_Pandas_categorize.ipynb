````python
drive_path = r"c:/Apps/temp/"

#사용자 아이디, 영화 아이디, 평점, ?
data = np.loadtxt(drive_path+"ratings_small.txt", delimiter="::", dtype = np.int64)
data
````

````python
#전체 평점 평균
avg_rating = data[:,2].mean() #np.mean(data[:, 2])
avg_rating
````

````python
#각 사용자별 평점 평균

score = []
user_id =  np.unique(data[:, 0])
for i in user_id:
    user_score_mean = data[data[:, 0]==i][:, 2].mean()
    score.append([i, user_score_mean])
    
score
````

````python
#1번 사용자 영화평점 평균 -> ?
first_user = data[data[:,0]==1]
first_user[:,2].mean()
````

````python
#평점평균 4이상인 사용자

over_4 = score[score[:,1]>=4]
over_4 = over_4[:, 0].astype('int')
over_4
````

### Categorizing

````python
#Categorzing
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 100]
bins = [0, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels = labels)
````
