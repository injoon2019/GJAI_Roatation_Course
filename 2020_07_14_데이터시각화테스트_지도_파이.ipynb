### Matplotlib 연습 01


```python
drive_path = r"c:/Apps/temp/"
data = pd.read_csv(drive_path+"Traffic_Accident_2017.csv", encoding="euc-kr")
data
```

```python
# 요일별 사고 건수
t = data["요일"].value_counts()
#value 순서대로 나온 것을 원하는 index 순서대로 정리하고 싶을 떄
t = t[['월', '화', '수', '목', '금', '토', '일']]
t
```

```python
x = range(7)
plt.plot(x, t)
plt.show()
```

```python
# x의 tick 바꾸기
# y의 limm 바꾸기 
x = range(7)
plt.xticks(x, ['월', '화', '수', '목', '금', '토', '일'])
plt.xlabel("요일")
plt.ylabel("사고 건수")
plt.bar(x, t)
plt.ylim(500, 620)
plt.title("요일별 교통사고 사망 건수")
plt.show()
```

```python
## 차vs차 사고 다발지역 (발생지 시도 시각화) : 발생지시도, 사고유형_대분류
data.T

#data[["발생지시도", "사고유형_대분류"]]
#t = data['사고유형_대분류'][data['사고유형_대분류']=='차대차']
t = data[["발생지시도", "사고유형_대분류"]][data['사고유형_대분류']=='차대차']
print(t)

t1 = t['발생지시도'].value_counts()
y = t1[['서울', '대구', '부산', '광주', '경기', '제주']]
x = range(6)
# 다른방법  
# x = t1.index()
# y = t1.counts()
plt.title('차 vs 차 사고 다발지역')
plt.xticks(x, ['서울', '대구', '부산', '광주', '경기', '제주'])
plt.bar(x, y)
plt.show()
```

```python
#교통사고 다발 시간대 시각화
# 0-2, 3-5, 9-11, 12-14, 15-17, 18-20, 21-23시 8개의 구간으로 나누어
def time_slice(x):
    return x%100

data["발생시간"] = data["발생년월일시"].apply(time_slice)

#이 부분바뀌었다. 강사님이준자료 참고해라 
bins = [-1,3,6,9,12,15,18,21,24]
labels = ["0-2", "3-5", "6-9", "9-11", "12-14", "15-17", "18-20", "21-24"]
time_cuts = pd.cut(data["발생시간"], bins, labels=labels)
time_cuts.value_counts()
#pd.cut(train['Age'], bins, labels = labels).value_counts()
#data.T


##강사님 답
# t = [x%100 for x in data['발생년월일시']]
# time_df = pd.DataFrame(t, columns=["사고시간"])

# time_df
```

```python
rs = time_cuts.value_counts()
plt.figure(figsize=(5,5))
plt.pie(rs, labels = labels, autopct = "%1.1f%%")
plt.show()
```

```python
# !pip install folium  --설치 명령어
import folium  #--지도 라이브러리

```

```python
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)  #지도의 중앙이되는 좌표
seoul_map
```

```python
#마커
m = data[["발생년월일시", "발생지시도", "경도", "위도"]][data["발생지시도"]=="서울"]
m.head(3)
```

```python
for name, lat, lng in zip(m['발생년월일시'], m["위도"], m["경도"]):  #묶어서 튜플로 제공
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
    
seoul_map.save("./seoul.html")
seoul_map
```
